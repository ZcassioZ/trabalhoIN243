```python
#Importar Bibliotecas
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import seaborn as sns
```

```python
# Importar dataset 1
dfc1 = pd.read_csv('https://raw.githubusercontent.com/lucheol/in243/master/projeto/cluster_dataset_1.csv')
dfc1.head()
```

```python
# Trabalhando dataset 1
scaler1 = StandardScaler()
scaled_1 = scaler1.fit_transform(dfc1)
scaled_1
dft1 = pd.DataFrame(scaled_1)
model = KMeans(n_clusters=15)
model.fit(scaled_1)
model.labels_
df_groups1 = pd.DataFrame(model.cluster_centers_, columns=dft1.columns)
df_groups1.transpose()
```

Plotar o gr√°fico apenas para ilustrar o peso de cada item nos Clusters

```python
df_groups1.transpose().plot.bar(subplots=True, figsize=(30,30), sharex=False)
```

```python
#Preparar o dataset para visualizar em 2D
tsne = TSNE()
visualizacao = tsne.fit_transform(scaled_1)
sns.set(rc={'figure.figsize': (10,10)})
sns.scatterplot(x=visualizacao[:,0], y=visualizacao[:,1], hue=model.labels_,
               palette=sns.color_palette('Set2',15))
```

```python
def get_kmeans(numero_clusters, df_groups1):
    model = KMeans(n_clusters=numero_clusters)
    model.fit(df_groups1)
    return [numero_clusters, model.inertia_]
```

```python
get_kmeans(15, scaled_1)
get_kmeans(3, scaled_1)
get_kmeans(40, scaled_1)
```

```python
resultado = [get_kmeans(numero_de_grupos, scaled_1) for numero_de_grupos in range(1,41)]
```

```python
resultado = pd.DataFrame(resultado, columns=['groups', 'inertia'])
```

```python
resultado.head()
```

```python
resultado['inertia'].plot(xticks=resultado['groups'])
```
